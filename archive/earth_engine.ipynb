```python
import rioxarray
import geopandas as gpd
rds = rioxarray.open_rasterio('geo_tiffs/temperature_annual_range_switzerland.tif')
rds.name = "data"
df = rds.squeeze().to_dataframe().reset_index()
geometry = gpd.points_from_xy(df.x, df.y)
annual_temp_range = gpd.GeoDataFrame(df, crs=rds.rio.crs, geometry=geometry)
```

```python
annual_temp_range.shape
```

```python
annual_temp_range.dropna(inplace=True)
annual_temp_range.shape
```

```python
annual_temp_range.head()
```

```python
annual_temp_range['data'].describe()
```
