```python
import rasterio as rio
from rasterio.plot import show
```

```python
# Load the tif file
fp = r'data/annual_mean_temperature_switzerland.tif'
raster = rio.open(fp)

# Plot the tif file
show(raster)
```

```python
# print rows and columns
print(f'Rows: {raster.height}')
print(f'Columns: {raster.width}')

#print coordinate value range
print(f'Coordinate Reference System: {raster.crs}')
print(f'Coordinate Reference System: {raster.bounds}')
```

```python
val = raster.read(1)[200, 100] # read the value at row 200, column 100

print(val)
```
